```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x = np.linspace(1,5,11)
x
y=x**3
y
```

```python
plt.plot(x,y)
plt.title('helo')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
```

**Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig=plt.figure(figsize=(20,8), dpi=100)
axis1 = fig.add_axes([0.1,0.1,0.4,0.8]) #[left, bottom, width, height].
axis1.plot(x,y)
axis2 = fig.add_axes([0.1,0.3,0.2,0.5])
axis2.plot(x,y)
```
